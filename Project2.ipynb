## New York City causes of death data

**Leading Causes of Death in New York City**

• What were the top 5 causes of death among Hispanics in 2007?

• What were the top 5 causes of death among Black Non-Hispanics in 2007?

• What were the top 5 causes of death among White-Non-Hispanics in 2007?

• What were the top 5 causes of death for Asian and Pacific Islanders in 2007?

*The CSV file can be found at the following link: https://github.com/lwaldron/IntroBiostat/blob/master/data/New_York_City_Leading_Causes_of_Death.csv *


```python
#the code below reads the csv file
deathdata = pd.read_csv('https://raw.githubusercontent.com/lwaldron/IntroBiostat/master/data/New_York_City_Leading_Causes_of_Death.csv')
```

```python
#The following code will return a table with all the results in the csv file
deathdata
```

```python
#The code below can be used to find the top cause of death for Hispanics in 2007:
deathdata[(deathdata['Year']==2007) & (deathdata['Ethnicity']=='HISPANIC')].sort_values('Count', ascending=False).head()
```

```python
#The code below can be used to find the top cause of death for NON-HISPANIC BLACK people in 2007:
deathdata[(deathdata['Year']==2007) & (deathdata['Ethnicity']=='NON-HISPANIC BLACK')].sort_values('Count', ascending=False).head()
```

```python
#The code below can be used to find the top cause of death for NON-HISPANIC White people in 2007:
deathdata[(deathdata['Year']==2007) & (deathdata['Ethnicity']=='NON-HISPANIC WHITE')].sort_values('Count', ascending=False).head()
```

```python
#The code below can be used to find the top cause of death for ASIAN & PACIFIC ISLANDERS in 2007:
deathdata[(deathdata['Year']==2007) & (deathdata['Ethnicity']=='ASIAN & PACIFIC ISLANDER')].sort_values('Count', ascending=False).head()
```

## California cities data

• Which city has the highest and lowest elevation? 

• Which city has the highest and lowest population? 

• Which city has the greatest total area?

*The CSV file can be found at the following link: 
https://raw.githubusercontent.com/jakevdp/PythonDataScienceHandbook/master/notebooks/data/california_cities.csv *

```python
#the code below reads the csv file
citydata = pd.read_csv('https://raw.githubusercontent.com/jakevdp/PythonDataScienceHandbook/master/notebooks/data/california_cities.csv')
```

```python
#The following code will return a table with all the results in the csv file
citydata
```

```python
#the code below displays the top 5 cities based on highest elevation
citydata[["city", "elevation_ft"]].sort_values(by=["elevation_ft"], ascending=False).head()
```

```python
#the code below displays the top 5 cities based on lowest elevation
citydata[["city", "elevation_ft"]].sort_values(by=["elevation_ft"], ascending=True).head()
```

```python
#the code below displays the top 5 cities based on highest population
citydata[["city", "population_total"]].sort_values(by=["population_total"], ascending=False).head()
```

```python
#the code below displays the top 5 cities based on lowest population
citydata[["city", "population_total"]].sort_values(by=["population_total"], ascending=True).head()
```

*The data above seems incorrect, according to Google the population for Pomona is actually 152,494 as of 2016*

```python
#the code below will correct the table above
citydata.loc[(citydata.city== "Pomona"), "population_total"] = 152494
```

```python
#the code below will display the updated top 5 cities based on lowest population
citydata[["city", "population_total"]].sort_values(by=["population_total"], ascending=True).head()
```

```python
#the code below will display the top 5 cities based on greatest total area 
citydata[["city", "area_total_sq_mi"]].sort_values(by=["area_total_sq_mi"], ascending=False).head()
```
